<h3>#1. Create a Hypothetical Bike Riding Dataset</h3>

<p>This code imports essential data analysis and visualization libraries and creates a synthetic dataset for analysis. Here's a step-by-step explanation:</p>

<ul>
  <li><code>import pandas as pd</code> - Imports the pandas library with the alias <b>pd</b>, used for data manipulation and analysis.</li>
  <li><code>import numpy as np</code> - Imports the numpy library with the alias <b>np</b>, used for numerical operations.</li>
  <li><code>import matplotlib.pyplot as plt</code> - Imports the pyplot module from matplotlib with the alias <b>plt</b>, used for creating visualizations.</li>
  <li><code>import seaborn as sns</code> - Imports the seaborn library with the alias <b>sns</b>, used for statistical data visualization.</li>
  <li><code>np.random.seed(42)</code> - Sets a random seed to ensure reproducibility of random operations.</li>
  <li>The code then creates a synthetic dataset with bike riding data including:
    <ul>
      <li>Dates</li>
      <li>Number of rides</li>
      <li>Temperature</li>
      <li>Humidity</li>
      <li>Windspeed values</li>
    </ul>
    using various random number generation functions.
  </li>
  <li>Finally, it creates a <b>pandas DataFrame</b> from this data for further analysis.</li>
</ul>


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
dates = pd.date_range(start="2025-01-01", periods=90, freq="D")

data = {
    "date": dates,
    "rides": np.random.poisson(lam=500, size=90) + np.random.randint(-50, 50, 90),
    "temperature": np.random.normal(loc=20, scale=5, size=90),  # mean=20Â°C
    "humidity": np.random.randint(40, 90, size=90),
    "windspeed": np.random.normal(loc=15, scale=5, size=90)
}

df = pd.DataFrame(data)
```

### # Correlation matrix
